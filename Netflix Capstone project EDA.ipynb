```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
pd.set_option('display.max_rows',None)
```

```python
df=pd.read_csv('netflix-rotten-tomatoes-metacritic-imdb.csv')
```

```python
df.head()
```

```python
df.drop(['Netflix Link','IMDb Link','Image','Summary','Poster','TMDb Trailer'],axis=1,inplace=True)
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df['Genre'].fillna('Comedy',inplace=True)
```

```python
df['Tags'].fillna('Dramas',inplace=True)
```

```python
df['Languages'].fillna('English',inplace=True)
```

```python
df['Hidden Gem Score'].fillna(8.0,inplace=True)
```

```python
df['Country Availability'].fillna('Japan',inplace=True)
```

```python
df['Runtime'].fillna('1-2 hour',inplace=True)
```

```python
df['Director'].fillna(method='ffill',inplace=True)
```

```python
df['Writer'].fillna(method='ffill',inplace=True)
```

```python
df['Actors'].fillna(method='ffill',inplace=True)
```

```python
df['View Rating'].fillna(method='ffill',inplace=True)
```

```python
df['IMDb Score'].fillna(6.4,inplace=True)
```

```python
df['Rotten Tomatoes Score'].fillna(method='ffill',inplace=True)
```

```python
df['Metacritic Score'].fillna(method='ffill',inplace=True)
```

```python
df['Awards Received'].fillna(method='ffill',inplace=True)
```

```python
df['Awards Nominated For'].fillna(method='ffill',inplace=True)
```

```python
df['Boxoffice'].fillna(method='ffill',inplace=True)
```

```python
df['Release Date'].fillna('12 Oct 2018',inplace=True)
```

```python
df['Production House'].fillna(method='ffill',inplace=True)
```

```python
df['Production House'].fillna(method='bfill',inplace=True)
```

```python
df['IMDb Votes'].fillna(method='ffill',inplace=True)
```

```python
df['Trailer Site'].fillna(method='ffill',inplace=True)
df['Trailer Site'].fillna(method='bfill',inplace=True)
```

```python
df.head()
```

```python
date_col=pd.DatetimeIndex(df['Release Date'])
df['Release date(Year)']=date_col.year
df['Release date(Month)']=date_col.month
```

```python
date_col=pd.DatetimeIndex(df['Netflix Release Date'])
df['Netflix Release Date(Year)']=date_col.year
df['Netflix Release Date(Month)']=date_col.month
df['Netflix Release Date(Day)']=date_col.day
```

```python
df.head()
```

```python
df['Boxoffice']=df['Boxoffice'].str.replace('[\$\,]',"").astype(int)
df.head()
```

```python
import calendar
df['Netflix Release Date(Month)'] = df['Netflix Release Date(Month)'].apply(lambda x: calendar.month_abbr[x])
```

```python
df.head()
```

```python
df1=df.sort_values(by='Boxoffice',ascending=False)
```

```python
df1.head()
```

# Netflix release date

```python
fig_dims = (20,10)
fig, ax = plt.subplots(figsize=fig_dims)
plots=df['Netflix Release Date(Day)'].value_counts().plot(kind='bar',color='crimson')
plt.xlabel('Netflix release date')
plt.ylabel('count')
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=8, xytext=(0, 8),
                   textcoords='offset points')

```

# Top 10 most successful genre

```python
fig_dims = (10,6)
fig, ax = plt.subplots(figsize=fig_dims)
plots=df1['Genre'].value_counts().nlargest(10).plot(kind='barh',color='maroon')
plt.xlabel('Count')
plt.ylabel('Genre')
```

```python
large10=df1.nlargest(10,'Boxoffice')
large10
```

# Most earned movies on the basis of actors and directos

```python
fig_dims = (15,10)
fig, ax = plt.subplots(figsize=fig_dims)
a=sns.barplot(x='Director',y='Boxoffice',data=large10,hue='Actors',color='firebrick',edgecolor='yellow')
a.set(xlabel='Actos & Directors', ylabel='Boxoffice Collection')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

# Total percentage movie and series released

```python
ax=df['Series or Movie'].value_counts().plot(kind='pie',autopct='%1.0f%%',textprops={'color':"white",'weight':'bold','fontsize':15},radius=1.5,colors = ['indigo', 'magenta'])
plt.title('Percentage of movie and series released', pad=32)
plt.legend()
ax.yaxis.set_visible(False)
```
